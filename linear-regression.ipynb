```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Get de data of the file data.csv
points = np.genfromtxt("data.csv", delimiter=",")
```

```python
# show the first 10 examples of the data
points[:10]
```

```python
X = np.array([]) # this will have all X elements
Y = np.array([]) # this will have all Y elements

for i in points: #insert x and y elements from points to X and Y arrays
    X = np.append(X, i[0])
    Y = np.append(Y, i[1])
    
X[:10]
```

```python
Y[:10]
```

```python
# showing the data in a graph

b = 0 # initial y-intercept guess
m = 0 # initial slope guess

line = 1*m + b

fig,ax = plt.subplots()
ax.plot(X, Y, "go") #Initialize the graph
ax.plot(X, m*X + b)
plt.show()
```

```python
learning_rate = 0.0001
num_iterations = 1000
```

```python
def cost(b, m, points):
    error = 0
    for i in range(0, len(points)):
        x = X[j]
        y = Y[j]
        error += (y - (m * x + b)) ** 2
    return (error / float(len(points)))
```

```python
for i in range(num_iterations):
    costo = cost(b, m, points)
    print("iteration: {}, cost: {}".format(i, costo))
    b_gradient = 0
    m_gradient = 0
    N = float(len(points))
    for j in range(0, len(points)):
        x = X[j]
        y = Y[j]
        b_gradient += -(2/N) * (y - ((m * x) + b))
        m_gradient += -(2/N) * x * (y - ((m * x) + b))

    b = b - (learning_rate * b_gradient)
    m = m - (learning_rate * m_gradient)

print("b: {}, m:{}".format(b, m))
```

```python
fig,ax = plt.subplots()
ax.plot(X, Y, "go") #Initialize the graph
ax.plot(X, m*X + b)
plt.show()
```
